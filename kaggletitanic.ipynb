```python
import numpy as np
import pandas as pd
train = pd.read_csv("http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/train.csv")
test = pd.read_csv("http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/test.csv")
print train.head()
print "--------------------------------------------------------------------------------------------"
print test.head()
print "--------------------------------------------------------------------------------------------"
print train.describe(),"-------------------", test.describe()
```

```python
print "Only survived: ", train["Survived"].value_counts()
print "As proportions: ", train["Survived"].value_counts(normalize=True)

print "Male survivers vs dead ", train["Survived"][train["Sex"]=="male"].value_counts()
print "Female survivers vs dead ", train["Survived"][train["Sex"]=="female"].value_counts()

print "Normalised male survivors", train["Survived"][train["Sex"]=="male"].value_counts(normalize=True)
print "Normalised female survivors", train["Survived"][train["Sex"]=="female"].value_counts(normalize=True)
```

```python
#Checking if age played a role
train["Child"]=float("NaN")

train["Child"][train["Age"]<18] = 1
train["Child"][train["Age"]>=18] = 0

print "Normalized survival rates for under 18", train["Survived"][train["Child"]==1].value_counts(normalize=True)
print "Normalized survival rates for 18+", train["Survived"][train["Child"]==0].value_counts(normalize=True)
```

```python
#Setting all females in test set = 1(survived) because as seen 74% of females survived in training data whereas only 18% males survived
#Create a copy of test

test_pred1 = test
#test doesnt have a survived column yet. initialise it to 0 then set it to 1 for all females
test_pred1["Survived"] = 0

test_pred1["Survived"][test_pred1["Sex"]=="female"] = 1
test_pred1["Survived"][test_pred1["Sex"]=="male"] = 0

print test_pred1["Survived"]
```

```python
#Cleanup the data
#Change "Sex" as female=1 and male=0
#Change 3 classes of Embarked as S=0 C=1 Q=2
train["Sex"][train["Sex"]=="female"] = 1
train["Sex"][train["Sex"]=="male"] = 0
j = np.where(pd.isnull(train["Embarked"]))
print j #Where Embarked is empty
train["Embarked"] = train["Embarked"].fillna("S") #Fill empty spot(s) with S thats most frequent
train["Age"] = train["Age"].fillna(train["Age"].median()) #Fill empty ages with median age
train["Embarked"][train["Embarked"] == "S"] = 0
train["Embarked"][train["Embarked"] == "C"] = 1
train["Embarked"][train["Embarked"] == "Q"] = 2

print train["Sex"]
print train["Embarked"]
```

```python
#Import tree from sklearn
from sklearn import tree
target = train["Survived"].values
features_one = train[["Pclass", "Sex", "Age", "Fare"]].values
my_tree_one = tree.DecisionTreeClassifier()

my_tree_one = my_tree_one.fit(features_one, target)
print "Feature importances: ", my_tree_one.feature_importances_
print "Score: ", my_tree_one.score(features_one, target)
```

```python
test["Fare"]=test["Fare"].fillna(test["Fare"].median())
test["Sex"][test["Sex"]=="female"] = 1
test["Sex"][test["Sex"]=="male"] = 0
test["Age"]=test["Age"].fillna(test["Age"].median())
#Use predict() method with model my_tree_one with features of test data

test_features1 = test[["Pclass", "Sex", "Age", "Fare"]].values

prediction1 = my_tree_one.predict(test_features1)
print "Number of predictions on test data:", prediction1.size #Check for 418 predictions 

#Now create this as a CSV file
#Save PassengerId and Prediction1 as arrays
PassengerId = np.array(test["PassengerId"]).astype(int)
solution1 = pd.DataFrame(prediction1, PassengerId, columns = ["Survived"])
solution1.to_csv("titanicsolution1.csv", index_label = ["PassengerId"])
print solution1
```

```python
#Theres likely overfitting as score is 0.72249
#Lets try fitting the training data with more features and specifications.

features_two = train[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked"]].values
my_tree_two = tree.DecisionTreeClassifier(max_depth=10, min_samples_split=5, random_state=1)
my_tree_two = my_tree_two.fit(features_two, target)
print "Features importances: ", my_tree_two.feature_importances_
print "Score: ", my_tree_two.score(features_two, target)
```

```python
test["Embarked"][test["Embarked"] == "S"] = 0
test["Embarked"][test["Embarked"] == "C"] = 1
test["Embarked"][test["Embarked"] == "Q"] = 2

test_features2 = test[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked"]].values
prediction2 = my_tree_two.predict(test_features2)
print prediction2.size
solution2 = pd.DataFrame(prediction2, PassengerId, columns=["Survived"])
solution2.to_csv("titanicsolution2.csv", index_label=["PassengerId"])
print solution2
#Kaggle score is now 0.76077
```

```python
#Create an extra feature family size in a copy of training data and test that on a 3rd tree
train_copy = train.copy()
train_copy["fam_size"] = train_copy["SibSp"]+train_copy["Parch"]+1
features_three = train_copy[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked", "fam_size"]].values
my_tree_three = tree.DecisionTreeClassifier(max_depth = 10, min_samples_split=5, random_state=1)
my_tree_three = my_tree_three.fit(features_three, target)
print "Features importances: ", my_tree_three.feature_importances_
print "Score: ", my_tree_three.score(features_three, target)
```

```python
test["fam_size"] = test["SibSp"]+test["Parch"]+1
test_features3 = test[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked", "fam_size"]].values
prediction3 = my_tree_three.predict(test_features3)
print prediction3.size
solution3 = pd.DataFrame(prediction3, PassengerId, columns=["Survived"])
solution3.to_csv("titanicsolution3.csv", index_label=["PassengerId"])
print solution3
#Kaggle score is now 0.76555
```
